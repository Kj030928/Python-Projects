```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation

print("2D heat equation solver")

plate_length = 50
max_iter_time = 750

alpha = 2
delta_x = 1

delta_t = (delta_x ** 2)/(4 * alpha)
gamma = (alpha * delta_t) / (delta_x ** 2)

# Initialize solution: the grid of u(k, i, j)
u = np.empty((max_iter_time, plate_length, plate_length))

# Initial condition everywhere inside the grid
u_initial = 0

# Boundary conditions
u_top = 100.0
u_left = 0.0
u_bottom = 0.0
u_right = 0.0

# Set the initial condition
u.fill(u_initial)

# Set the boundary conditions
u[:, (plate_length-1):, :] = u_top
u[:, :, :1] = u_left
u[:, :1, 1:] = u_bottom
u[:, :, (plate_length-1):] = u_right

def calculate(u):
    for k in range(0, max_iter_time-1, 1):
        for i in range(1, plate_length-1, delta_x):
            for j in range(1, plate_length-1, delta_x):
                u[k + 1, i, j] = gamma * (u[k][i+1][j] + u[k][i-1][j] + u[k][i][j+1] + u[k][i][j-1] - 4*u[k][i][j]) + u[k][i][j]

    return u

def plotheatmap(u_k, k):
    # Clear the current plot figure
    plt.clf()

    plt.title(f"Temperature at t = {k*delta_t:.3f} unit time")
    plt.xlabel("x")
    plt.ylabel("y")

    # This is to plot u_k (u at time-step k)
    plt.pcolormesh(u_k, cmap=plt.cm.jet, vmin=0, vmax=100)
    plt.colorbar()

    return plt

# Do the calculation here
u = calculate(u)

def animate(k):
    plotheatmap(u[k], k)

anim = animation.FuncAnimation(plt.figure(), animate, interval=1, frames=max_iter_time, repeat=False)
anim.save("heat_equation_solution.gif")

print("Done!")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def heat_conduction(length, area, conductivity, temperature_initial, temperature_external, num_points=100):
    """
    Solve steady-state conduction through a wall.

    Parameters:
    - length: Thickness of the wall (meters)
    - area: Cross-sectional area of the wall (square meters)
    - conductivity: Thermal conductivity of the material (W/(m*K))
    - temperature_initial: Initial temperature on one side of the wall (Celsius)
    - temperature_external: External temperature on the other side of the wall (Celsius)
    - num_points: Number of points to discretize the wall

    Returns:
    - temperatures: Array of temperatures at each point through the wall
    - positions: Array of positions corresponding to each temperature point
    """
    dx = length / (num_points - 1)
    positions = np.linspace(0, length, num_points)
    temperatures = np.zeros_like(positions)

    # Initial conditions
    temperatures[0] = temperature_initial
    temperatures[-1] = temperature_external

    # Finite difference method for steady-state conduction
    for i in range(1, num_points - 1):
        temperatures[i] = temperatures[i - 1] + (temperature_external - temperature_initial) * dx / length

    return temperatures, positions

def plot_temperature_distribution(positions, temperatures):
    plt.plot(positions, temperatures, marker='o')
    plt.xlabel('Position (meters)')
    plt.ylabel('Temperature (Celsius)')
    plt.title('Temperature Distribution in a Wall')
    plt.grid(True)
    plt.show()

# Example usage
length_wall = 0.1  # meters
area_wall = 0.01  # square meters
thermal_conductivity = 10.0  # W/(m*K)
initial_temp = 100.0  # Celsius
external_temp = 25.0  # Celsius

temperatures, positions = heat_conduction(length_wall, area_wall, thermal_conductivity, initial_temp, external_temp)
plot_temperature_distribution(positions, temperatures)
```
