```python
def add(num1, num2):
    a = num1 + num2
    return a

def sub(num1, num2):
    s = num1 - num2
    return s

def mult(num1, num2):
    m = num1*num2
    return m

def divi(num1, num2):
    d = num1/num2
    return d

def simple_calculator():
    operations = ['+', '-', 'x', '/']
    
    user_input = input("Enter an operation: ")
    
    if user_input not in operations:
        print('Please enter a valid operation')
        return 

    try:
        num1 = int(input("Enter the first number: "))

    except ValueError:
        print("It must be a valid number")
    

    try:
        num2 = int(input("Enter the second number: "))

    except ValueError:
        print("It must be a valid number")


    if user_input == "+":
        print(add(num1, num2))
    
    elif user_input == "-":
        print(sub(num1, num2))
    
    elif user_input == "x":
        print(mult(num1, num2))
    
    elif user_input == "/":
        try: 
            print(divi(num1, num2))
        except:
            print('Zero division is not allowed')
    else:
        print('Please enter a valid number or operation')
        
simple_calculator() 
```
